I want to find out how many episodes of Spongebob Squarepants the charcter Patrick Star appears in.

```python
import pandas as pd
```

```python
df = pd.read_csv('/content/spongebob_episodes.csv')
```

```python
if 'characters' not in df.columns or 'title' not in df.columns:
    print("Error: 'characters' or 'title' column not found. Please check the dataset.")
    exit()
```

```python
def process_character_string(char_str):
    try:
        if isinstance(char_str, list): # Already a list, return as is
          return char_str
        if pd.isna(char_str) or not isinstance(char_str, str):  # handle NaN or other non-string values
            return []
        char_str = char_str.replace('[', '').replace(']', '') #Remove brackets
        char_list = [s.strip("' ") for s in char_str.split(',')] #Remove quotes and spaces
        return [s for s in char_list if s] #filter out empty strings
    except Exception as e:
      print(f"Error processing character string: {char_str}, Error: {e}")
      return [] # Return an empty list to avoid breaking the rest of the code.


df['characters'] = df['characters'].apply(process_character_string)
```

```python
patrick_episodes = df[df['characters'].apply(lambda x: 'Patrick Star' in x)]
num_patrick_episodes = len(patrick_episodes)

print(f"Patrick Star appears in {num_patrick_episodes} episodes.")
```

```python
print("\nEpisodes with Patrick Star:")
print(patrick_episodes['title'])
```

It looks like Patrick Star appears in 434 episodes out of the 583 total. Thats quite a bit!
